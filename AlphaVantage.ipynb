# Lendo o JSON

```python
import requests
import pandas as pd

API_URL = "https://www.alphavantage.co/query"

data = {
    "function": "TIME_SERIES_DAILY_ADJUSTED",
    "symbol": 'BBSE3.SA',
    "outputsize": "full",
    "datatype": "json",
    "apikey": "1VI1L6694V2Q3CFA",
}

response = requests.get(API_URL, params=data)

info = response.json()
```

```python
try:
    print(info['Time Series (Daily)']['2019-12-01'])
except KeyError as ke:
    print('nao tem info para essa data ' + str(ke))
```
