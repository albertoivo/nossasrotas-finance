# Acessando a API

```python
from alpha_vantage.timeseries import TimeSeries
from alpha_vantage.techindicators import TechIndicators
import matplotlib.pyplot as plt
```

```python
ts = TimeSeries(key='1VI1L6694V2Q3CFA', output_format='pandas')

intraday, meta_intra = ts.get_intraday(symbol='ITUB4.SA', interval='5min', outputsize='compact')
daily, meta_daily = ts.get_daily('ITUB4.SA', outputsize='compact')
```

```python
intraday['4. close'].plot(figsize=(17,6));
```

```python
daily['4. close'].plot(figsize=(17,6));
```
