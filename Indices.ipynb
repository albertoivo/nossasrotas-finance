## Dataframes por Índice

## Normalizando todos os índices

```python
indices_adj_close = pd.read_csv('csv/indices_adj_close.csv', index_col=0)
(indices_adj_close / indices_adj_close.iloc[0] * 100).plot(figsize=(17,8));
```

## Normalizando os índices sem o IBOV

```python
indices_ex_adj_close = pd.read_csv('csv/indices_ex_adj_close.csv', index_col=0)
(indices_ex_adj_close / indices_ex_adj_close.iloc[0] * 100).plot(figsize=(17,8));
```

## Taxa de retorno diária dos índices

```python
ind_returns = (indices_adj_close / indices_adj_close.shift(1)) - 1
ind_returns
```

## Taxa de retorno anual dos índices

```python
annual_ind_returns = ind_returns.mean() * 250 # média de pregões no ano
round(annual_ind_returns * 100, 3)
```
